```python
y = rb_exp3_5['Rnd']
x = rb_exp3_5.drop(['Rnd', 'Pick'], axis=1)
X = sm.add_constant(x)

lm = LinearRegression() 
lm01 = lm.fit(X,y)
pred01 = lm01.predict(X)

plt.figure(figsize=(7,7))
sns.scatterplot(x=pred01, y= y)
plt.title("Predicted vs. Actual Rnd", fontsize = 15)
plt.xlabel("Predicted Rnd (Yhat)", fontsize = 12)
plt.ylabel("Actual Rnd (Y)", fontsize = 12)
plt.axis([0,7,0,7])
plt.plot([0,7],[0,7], color = 'purple')  # 45 degree line
plt.grid()

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.80, random_state=2)
eq01 = lm.fit(X_train,y_train)
accuracies_eq01 = cross_val_score(estimator=lm, X=X_train, y=y_train, cv=10, n_jobs= -1)
eq01_accy = accuracies_eq01.mean()
print(round((eq01_accy),5))

rig = linear_model.Ridge(alpha=0.01)  # Laregr the alpha, the simpler the model. More bias & less variance.
rig.fit(X_train, y_train)
y_train_pred = rig.predict(X_train)
y_test_pred = rig.predict(X_test)
print("Ridge Train RMSE: ", np.sqrt(metrics.mean_squared_error(y_train, y_train_pred)))
print("Ridge Test RMSE: ", np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))  
print("Ridge Coef: ", rig.coef_)

las = linear_model.Lasso(alpha=0.01)
las.fit(X_train, y_train)
y_train_pred = las.predict(X_train)
y_test_pred = las.predict(X_test)
print("Lasso Train RMSE: ", np.sqrt(metrics.mean_squared_error(y_train, y_train_pred)))
print("Lasso Test RMSE: ", np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))  
print("Lasso Coef: ", las.coef_)
```
